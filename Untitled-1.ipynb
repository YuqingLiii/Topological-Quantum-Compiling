```python
import numpy as np
import matplotlib.pyplot as plt

# 定义固定的epsilon和phi
epsilon = 0.1  # 假设epsilon = 0.1，根据需要调整
phi = np.pi / 4  # 假设phi = pi/4，根据需要调整

# 定义函数f(u)
def f(u):
    return np.sqrt(1 - np.abs(np.real(u * np.exp(1j * phi / 2)))) - epsilon

# 定义u的范围
u_real = np.linspace(-2, 2, 400)
u_imag = np.linspace(-2, 2, 400)
U_real, U_imag = np.meshgrid(u_real, u_imag)
U = U_real + 1j * U_imag

# 计算f(u)的值
F = f(U)

# 绘制等高线图
plt.figure(figsize=(8, 6))
contour = plt.contour(U_real, U_imag, F, levels=[0], colors='blue')
plt.clabel(contour, inline=True, fontsize=8)
plt.xlabel('Re(u)')
plt.ylabel('Im(u)')
plt.title(r'$\sqrt{1 - |\Re(u e^{i\phi/2})|} \leq \epsilon$')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

```python
def custom_mod5(n):
        remainder = n % 5
        if remainder == 3:
            return -2
        elif remainder == 4:
            return -1
        else:
            return remainder
```

```python
def binary_gcd( a, b, log_file="gcd_log__.txt"):
    # 打开文件进行写操作，如果文件不存在则会创建
        
        with open(log_file, "a", encoding="utf-8") as f:
            def log_message(message):
                f.write(message +"\n")
            
            log_message("本次的a，b是：")
            log_message(a.get_str())
            log_message(b.get_str())
        
            """
                在 Z[ω] 环上实现 BINARY-GCD 算法
            """
            #print("本次的a，b是：")
            #print(a.get_str())
            #print(b.get_str())
            if a.get_value() == 0:
                return b
            if b.get_value() == 0:
                return a
            log_message(f"is_Divide(a): {is_Divide(a)}, is_Divide(b): {is_Divide(b)}")
            #print(f"is_Divide(a): {is_Divide(a)}, is_Divide(b): {is_Divide(b)}")
            if is_Divide(a) == True and is_Divide(b) == True:
                a1 = factor_1_plus_omega(a)
                b1 = factor_1_plus_omega(b)
                return RingCycIntegers([1,1,0,0]).mul(binary_gcd(a1, b1, log_file))

            if is_Divide(a) == True and is_Divide(b) == False:
                a1 = factor_1_plus_omega(a)
                log_message(f"a可以分且a1是{a1.get_str()}\n")
                return binary_gcd(a1,b, log_file)
            
            if is_Divide(a) == False and is_Divide(b) == True:
                b = factor_1_plus_omega(b)
                
                while is_Divide(b) == True:
                    b = factor_1_plus_omega(b)
                log_message(f"b可以分且b是{b.get_str()} \n")
                return binary_gcd(a,b, log_file)
                
            u = v = RingCycIntegers([1,0,0,0])
            if is_Divide(a) == False and is_Divide(b) == False:
                
                a_mod = a.get_coeff(0) - a.get_coeff(1) + a.get_coeff(2) - a.get_coeff(3)
                if a_mod % 5 == 2:
                    u = RingCycIntegers([2, 0, 0, 0])
                elif a_mod % 5 == 3:
                    u = RingCycIntegers([-2, 0, 0, 0])
                else:
                    u = custom_mod5(sp.mod_inverse(a_mod, 5))
                    u = RingCycIntegers([u, 0, 0, 0])
                    
                #u = self.custom_mod5(sp.mod_inverse(a_mod, 5))
                #u = RingCycIntegers([u, 0, 0, 0])
                b_mod = b.get_coeff(0) - b.get_coeff(1) + b.get_coeff(2) - b.get_coeff(3)
                if b_mod % 5 == 2:
                    v = RingCycIntegers([2, 0, 0, 0])
                elif b_mod % 5 == 3:
                    v = RingCycIntegers([-2, 0, 0, 0])
                else:
                    v =custom_mod5(sp.mod_inverse(b_mod, 5))
                    v = RingCycIntegers([v, 0, 0, 0])
                #v =self.custom_mod5(sp.mod_inverse(b_mod, 5))
                #v = RingCycIntegers([v, 0, 0, 0])
                #u=self.mod_inverse(a)
                #v=self.mod_inverse(b)
                log_message("计算出来的逆元都是：")
                log_message(f"{u.get_str()}, {v.get_str()}")
                #print("计算出来的逆元都是：")
                #print(f"{u.get_str()}, {v.get_str()}")
            log_message("a和b的模大小是：")
            log_message("   ")
            log_message(str(a.abs_quadratic().get_value()))
            log_message("   ")
            log_message(str(b.abs_quadratic().get_value()))
            log_message(" \n  ")
            if a.abs_quadratic().get_value() < b.abs_quadratic().get_value():
                c = a
            else:
                c = b

            return binary_gcd(c, u.mul(a).sub(v.mul(b)), log_file)
```

```python
from TwoRings import SubRealRing,RingCycIntegers
def unit_dlog(u):
        """实现 UNIT-DLOG 算法"""
        
        a, b = u.get_coeff(0),u.get_coeff(1)
        
        s = 1
        k = 0
        if a < 0:
            a = -a
            b = -b
            s = -s
        mu = a * b
        while abs(mu) > 1:
            #print(a,b,mu)
            
            with open("unit_dlog", "a", encoding="utf-8") as f:
                f.write(str(a))
                f.write("   ")
                f.write(str(b))
                f.write("   ")
                f.write(str(mu))
                f.write(" \n  ")
            
            temp=a-b
            if mu>1:
                a=b
                b=temp
                k -= 1
            else:
                #a, b = a, temp
                a=a
                b=temp
                k += 1
            mu = a * b
```

```python
from TwoRings import SubRealRing,RingCycIntegers
def unit_dlog(u):
        """实现 UNIT-DLOG 算法"""
        tau=self.__class__.tau
        a, b = extract_a_b(u)

        s = 1
        k = 0
        if a < 0:
            a = -a
            b = -b
            s = -s
        mu = a * b
        while abs(mu) > 1:
            temp=a-b
            if mu>1:
                a, b = b, temp
                k -= 1
            else:
                a, b = a, temp
                k += 1
            mu = a * b

        v_values = [1, -1, tau, -tau, tau**2, -tau**2]
        for v in v_values:
            if sp.simplify(u - s * v * tau**k) == 0:                    
                if v == -1:
                    s = -s
                elif v == tau:
                    k += 1
                elif v == -tau:
                    s = -s
                    k += 1
                elif v == tau**2:
                    k += 2
                elif v == -tau**2:
                    s = -s
                    k += 2
                return s, k
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义 p 的范围
p_values = np.linspace(0, 1, 500)

# 定义 c* 和 q* 函数
def c_star(p):
    if 0 <= p <= 1/2:
        return 1 - p**2
    elif 1/2 < p <= 1:
        return -p**2 + 2*p

def q_star(p):
    if 0 <= p <= 1 - 1/np.sqrt(2):
        return 1 - p**2
    elif 1 - 1/np.sqrt(2) < p <= 1/np.sqrt(2):
        return 1/np.sqrt(2) * (1 - 2*p*(1 - p)) + 1/2
    elif 1/np.sqrt(2) < p <= 1:
        return -p**2 + 2*p

# 计算 c* 和 q* 的值
print(q_star(1-1/np.sqrt(2)))
print(c_star(1-1/np.sqrt(2)))
q_star_values = [q_star(p)-c_star(p) for p in p_values]

# 绘制图像
plt.figure(figsize=(10, 6))
plt.plot(p_values, q_star_values, label="q*")
plt.xlabel("p")
plt.ylabel("Value")
plt.title("q* and c* as functions of p")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

# 定义向量 v
v = np.array([0.309016994374947, 0.951056516295154])

# 定义基向量 b1, b2, b3
b0 = np.array([1,0])
b1 = np.array([sp.cos(sp.pi/5), sp.sin(sp.pi/5)])
b2 = np.array([sp.cos(2*sp.pi/5), sp.sin(2*sp.pi/5)])
b3 = np.array([sp.cos(3*sp.pi/5), sp.sin(3*sp.pi/5)])

# 构造矩阵 B
B = np.column_stack([b0,b1, b2, b3]).astype(float)

# 解线性方程 Bc = v
c = np.linalg.lstsq(B, v, rcond=None)[0]

# 打印结果
print("系数向量 c:", c)

# 验证结果
v_reconstructed = np.dot(B, c)
print("重构的向量 v:", v_reconstructed)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# 定义符号变量
u_real, u_imag, epsilon, phi = sp.symbols('u_real u_imag epsilon phi')

# 定义复数 u 和 exp(i*phi/2)
u = u_real + sp.I * u_imag
exp_term = sp.exp(sp.I * phi / 2)

# 定义不等式
inequality = sp.sqrt(1 - abs(sp.re(u * exp_term))) <= epsilon

# 设置 epsilon 和 phi 的值
epsilon_val = 0.1
phi_val = np.pi / 4

# 创建数值函数
u_real_vals = np.linspace(-2, 2, 400)
u_imag_vals = np.linspace(-2, 2, 400)
u_real_mesh, u_imag_mesh = np.meshgrid(u_real_vals, u_imag_vals)

inequality_func = sp.lambdify((u_real, u_imag, epsilon, phi), inequality, 'numpy')
inequality_vals = inequality_func(u_real_mesh, u_imag_mesh, epsilon_val, phi_val)

# 创建一个布尔掩码，显示满足不等式的区域
mask = inequality_vals

# 绘制复平面上的区域
plt.figure(figsize=(10, 8))
plt.contourf(u_real_mesh, u_imag_mesh, mask, levels=[0, 1], colors=['blue'], alpha=0.5)
plt.colorbar(label='Satisfy Inequality')
plt.xlabel('Re(u)')
plt.ylabel('Im(u)')
plt.title(r'$\sqrt{1 - |\Re(u e^{i\phi/2})|} \leq \epsilon$ with $\epsilon = 0.1$ and $\phi = \pi/4$')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```
